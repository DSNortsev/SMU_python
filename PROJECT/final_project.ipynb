```python
import pandas as pd

df = pd.read_csv('Oct1_2019_to_Sep30_2020.csv')
df_public_work = df.loc[df['Department'] == 'Public Works']
df_public_work
```

```python
df_public_work.shape
```

```python
# Find all missing/NaN values
df_public_work.isnull().sum(axis = 0)
```

```python
# Replace NaN with most frequent value
df_public_work = df_public_work.fillna(df_public_work.mode().iloc[0])
df_public_work.isnull().sum(axis = 0)
```

```python
# Show unique outcomes
df_public_work['Outcome'].unique()
```

```python
df_public_work[df_public_work['Outcome'] == 'Work Completed'].shape
```

```python
df_public_work[df_public_work['Outcome'] == 'Work Not Completed'].shape
```

```python
# Get total requests per Outcome
total = df_public_work.shape[0]
result = []
for outcome in df_public_work['Outcome'].unique():
    total_per_outcome = df_public_work[df_public_work['Outcome'] == outcome].shape[0]
    result.append([outcome, total_per_outcome, total_per_outcome/total])

# Sort List in descending order
result.sort(key=lambda x: x[1], reverse=True)

for outcome in result:
    print(f'Outcome: {outcome[0]:<50} Total: {outcome[1]:<5}  Total_percent: {outcome[2]:.2f}')
```

```python
# List Services per Outcome 
for outcome in result:
    services = df_public_work[df_public_work['Outcome'] == outcome[0]]['Service Request Type'].unique()
    print(f'Outcome: {outcome[0]}\n Services: {services}\n')
```

```python
df_public_work['Status'].unique()
```

```python
total = df_public_work.shape[0]
for status in df_public_work['Status'].unique():
    total_per_status = df_public_work[df_public_work['Status'] == status].shape[0]
    print(f'Outcome: {status:<23} Total: {total_per_status:<5}  Total_percent: {total_per_status/total:.2f}')
```

```python
# List all services in Progress
df_in_progress = df_public_work[df_public_work['Status'] == 'In Progress']
df_in_progress['Service Request Type'].value_counts()
```

```python
# List the method of communication
df_communication = df_public_work['Method Received Description'].value_counts()
total = df_public_work.shape[0]

for indx in df_communication.index.values:
    print(f'{indx:<9} - {df_communication.loc[indx]:<5} - {df_communication.loc[indx]/total:.2f}')
```

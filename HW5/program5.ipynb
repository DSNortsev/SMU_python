### 1)	Find a CSV data set online that has both categorical and numerical values in it

```python
import pandas as pd

df = pd.read_csv('insurance.csv')
df
```

### 2)	Programmatically evaluate your code for NaNs.  If there are any, replace them with meaningful values.

```python
df.isna().sum() 
```

```python
df.dtypes
```

```python
X_numerical = df[['age', 'bmi', 'children']].copy(deep=True)
X_categorical = df[['sex', 'smoker', 'region']].copy(deep=True)
y = df['charges'].copy(deep=True)
```

```python
X_numerical
```

```python
X_categorical
```

```python
y
```

```python
X_dummy = pd.get_dummies(X_categorical)
X_dummy.head()
```

```python
X = pd.concat([X_numerical, X_dummy], axis=1)
X.head()
```

### 3)	Run a regression.  Interpret your results.  You may either interpret your results in your code or you may include an accompanying Word or TXT document.

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression().fit(X,y)
```

```python
print(f'Intercept: {model.intercept_:>13.2f}\n')
for column, coef in zip(X.columns, model.coef_):
    print(f'{column:>16}: {coef:.2f}')
```
